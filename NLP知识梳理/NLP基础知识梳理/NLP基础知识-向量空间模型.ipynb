### TextRank表示

- 部分把词序信息和词之间的关系搬回文档

> 词和词之间的联系用图表征

> 一个词的分数可以用邻居表征

> 基于图的排序算法

- 具体方法

> 将文本分割成若干组成单元（词/句子）

> 将每个单元作为图的节点

> 用单元之间的相似度（单词时可以是共现次数或是否共现）作为边的权重

> 通过循环迭代计算单元的TextRank值，直到收敛

## 主成分分析PCA

- 基本概念

> 减少数据集的维数，同时保留数据集中对方差贡献最大的特征

> 主成分可以看作一个线性方程

- 基本思路

> 描述数据的主要特征是什么

> 互相独立的特征有哪些

> 消除特征之间的相关性

- 通用方法总结

> 计算协方差矩阵

> 计算特征值和特征向量

> 特征值从大到小排，对应的特征向量，形成的矩阵，即为最终结果
