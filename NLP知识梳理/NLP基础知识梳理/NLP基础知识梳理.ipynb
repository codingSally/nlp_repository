# 语料


- 语料是NLP的生命之源，所有的NLP问题都是从语料中学到数据分布的规律

## 语料分类

- 单语料

> 只有句子和句子的集合，如整理好的英文文档、中文古诗数据集等

- 平行语料

> 有句子和句子之间的 1 vs 1 对应关系，如中英文翻译数据集，对话数据集

- 其他

> 比较复杂的结构，如知乎的回答和评论数据
