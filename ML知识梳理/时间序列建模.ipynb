# 时间序列的概述

## 定义

- 根据已有的时间序列数据预测未来的变化

- 假设事物发展趋势会延伸到未来（会继续发生）

- 预测所依据的数据具有不规则性（未来有很多不确定性）

- 不考虑事物发展之间的因果关系（根据当前考虑未来


## 分类

- 时间序列分为平稳序列和非平稳序列两大类。

> 平稳序列是不存在趋势只存在随机性的序列

> 非平稳序列则是包含趋势、季节性和随机性的序列。

> - 趋势性：时间序列在长时间内呈现出来的长期上升或下降的变动；

> - 季节性：时间序列在一年内出现的周期性波动，比如说航空业的销售淡季和销售旺季；

> - 随机性：时间序列的偶然性波动。

## 平稳性检验

- 平稳性就是要求经由样本时间序列所得到的拟合曲线在未来一段时间内仍能顺着现有的形态惯性地延续下去。平稳性要求序列的均值和方差不发生明显变化。


> 严平稳

> - 序列随着时间的推移，其统计性质保持不变。如：白噪声（正太分布），期望为0，方差为1

> 宽平稳

> - 时间序列的均值和方差为与时间无关的常数；

> - 其协方差与时间间隔有关但与时间无关。

> 非平稳

> - 样本时间序列的均值、方差、协方差不是常数

> - 当前时间的均值、方差、协方差并非未来时间的均值、方差、协方差

## 差分法

- 使用差分法可以使得数据更平稳；常用的方法就是一阶差分法和二阶差分法。

## ARIMA模型

### 自回归模型(AR)

- 基础概念

![title](ml-img/ml16.png)

> 自相关系数是我们要求解的

- 缺点

> 自回归模型是用自身的数据来进行预测的

> 必须具有平稳性

> 必须具有自相关性，如果自相关系数($φ_{i}$)小于0.5，则不宜采用

> 自回归只适用于预测与自身前期相关的现象

### 差分自回归移动平均模型(ARIMA(p,d,q))

- Autoregressive Integrated Moving Average Model,

- AR是自回归，p为自回归项；MA为移动平均；q为平均项数，d为时间序列成为平稳时所做的差分次数

- 原理：将非平稳时间序列转化为平稳时间序列然后将因变量仅对它的滞后值以及随机误差项的现值和滞后值进行回归所建立的模型
